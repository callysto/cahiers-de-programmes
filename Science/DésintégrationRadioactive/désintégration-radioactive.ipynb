# Désintégration radioactive


## Half Life et Radioactif Decay


La demi-vie d'un atome instable est la quantité de temps requise en moyenne pour que la moitié d'une population de cet atome se désintègre en un élément différent. La demi-vie d'un atome est généralement mesurée en secondes. La demi-vie peut aller de yottoseconds ($10^{24}$ secondes!) À des nombres si importants qu'il est difficile de considérer l'atome instable. Par exemple, Tellerium-128 a une demi-vie d'environ années $10^{24}$ (yotta), comme dans toutes ces années. Une courte liste de demi-vies atomiques sélectionnées est disponible sur Wikipedia [ici](https://en.wikipedia.org/wiki/List_of_radioactive_isotopes_by_half-life).

Que signifie la demi-vie d'un atome? Supposons que vous ayez une collection de mille atomes avec une demi-vie de vingt minutes. Au bout de vingt minutes, votre collection de ces atomes aura diminué à environ cinq cents atomes. Cela ne signifie pas que vos atomes ont disparu! Vos atomes instables ont **décomposé**. Ce qui signifie que certains atomes ont perdu de l'énergie en émettant des radiations et ont changé leur nombre de neutrons, de protons ou les deux! Le plus souvent, les atomes se désintègrent en émettant un électron, un neutron, un positron ou une particule alpha (hélium sans électrons (He$^{2+}$)). Ce qui se passe dans chacun de ces processus de désintégration est décrit ci-dessous.



## Modes de décomposition

### Alpha ($\alpha$) Decay

La désintégration alpha est un processus de désintégration dans lequel un atome libère une particule $\alpha$. Une particule $\alpha$ est un autre nom pour un atome d'hélium ionisé $\left(^4_2He\right)$. Si un élément subit une désintégration alpha, l'atome réduit à la fois son nombre de neutrons $N$ et son nombre de protons $Z$. Un atome X avec un numéro atomique $A$, la somme des neutrons et des protons, se décomposant en un atome $Y$ via $\alpha$ decay a la forme suivante:


\begin{equation}
^A_ZX^N \rightarrow ^{A-4}_{Z-2}Y^{N-2} + ^4_2\alpha^2 
\end{equation}

Ici $A$ est le nombre atomique **** d'un atome, ou la somme de son nombre de neutrons $N$ et de protons $Z$.

\begin{equation}
A = Z + N
\end{equation}

La désintégration alpha, écrite comme exemple concret, la décomposition mathématique de l'uranium 235 apparaîtrait comme

\begin{equation}
_{92} ^{235} \text{U}^{143} \rightarrow _{90} ^{231}\text{Th}^{141} + _2 ^4\alpha^{2}
\end{equation}

On voit que l'uranium 235 se désintègre en Thorium 231 en émettant une particule $\alpha$ via la désintégration alpha. Notez que le nombre total de neutrons et de protons est égal à la fois du côté droit et du côté gauche de la réaction. Le nombre de neutrons n'est généralement pas écrit explicitement et calculé en soustrayant $Z$ de $A$.

### Beta $\beta$ Decay

$\beta$ decay se décline en deux saveurs: la décroissance bêta positive ($\beta^+$) et la décroissance bêta moins ($\beta^-$). Pendant $\beta^-$ decay, un atome émet un électron. Au cours de $\beta^+$, la décomposition d'un atome émet l'anti-particule de l'électron: le positron (une particule avec la même masse, mais la charge opposée d'un électron). Dans le cas de la désintégration $\beta^+$, l'atome émet un positon à partir d'un proton, ce qui signifie que l'atome «perd» un proton et «gagne» un neutron. Dans le cas de la désintégration $\beta^-$, un neutron émet un électron, ce qui signifie que l'atome "perd" un neutron et "gagne" un proton.

Ils émettent également des particules mystérieuses et communes connues sous le nom de neutrino

### Beta Minus pourriture

En règle générale mathay3 decay a la forme suivante

\begin{equation}
^A _Z\text{X}^N \rightarrow _{Z+1}^A\text{Y}^{N-1} + e^- + \bar{\nu}
\end{equation}

où $\bar{\nu}$ est un anti-neutrino. Notez que la charge et la masse sont conservées. De la formule ci-dessus, nous voyons qu'un neutron émet un électron et que l'anti-neutrino devient un proton. Remarquez comment la charge et la masse sont conservées. Nous avons obtenu une charge positive avec le proton, mais avons également obtenu une charge négative dans le processus avec un électron.

### Beta plus décroissance

\begin{equation}
^A _Z\text{X}^{N} \rightarrow _{Z-1}^A\text{Y}^{N+1} + e^+ + \nu
\end{equation}

où $\nu$ est un neutrino. Notez que la charge et la masse sont à nouveau conservées. De la formule ci-dessus, nous voyons qu'un proton émet un positron et un neutrino, et devient un neutron.

## Quelques fondements mathématiques
Faisons suite à notre discussion précédente et commençons à introduire une certaine rigueur mathématique dans notre discussion sur la demi-vie. La demi-vie d'un atome est notée $\tau_{1/2}$, mesurée en secondes, où $\tau$ est la lettre grecque tau. En utilisant la demi-vie d'un atome, on peut calculer le nombre d'atomes en fonction du temps $N(t)$ avec la relation suivante

\begin{equation}
N(t) = N_0 \; 2.718^{-\lambda t} 
\label{eq:decay-fr}
\tag{1}
\end{equation}

Où $N(t)$ est le nombre __ d'atomes que vous avez au moment $t$, $N_0$ est le nombre original d'atomes à $t=0$ et $\lambda$ est connu comme la "constante de demi-vie" qui a des unités de secondes inverses. La constante de demi-vie est un peu une abstraction de la demi-vie plus intuitive, cependant $\lambda$ est liée à la demi-vie par la relation simple suivante

\begin{equation}
\lambda = \frac{0.693}{\tau_{1/2}}
\label{eq:lambda-fr}
\tag{2}
\end{equation}

Les deux équations ci-dessus permettent de calculer la quantité d'atomes restant à un moment donné. Bien entendu, la demi-vie et le nombre d'atomes d'origine sont connus. Un fait plus subtil sur l'équation \ref{eq:decay-fr} est que la partie exponentielle ($2.718^{\lambda t}$) de cette équation est directement liée à la probabilité _qu'un atome décroît à un moment donné. Intuitivement, cela a du sens, que nous illustrons par un exemple. Supposons que nous ayons commencé avec mille atomes et après dix secondes, nous trouvons qu'il ne reste que 60% de nos atomes. En utilisant l'équation \ref{eq:decay-fr}, nous pouvons écrire ce scénario comme suit avec $N_0 = 1000$

\begin{equation}
N(t= 3 \text{ s}) = 1000 \times 2.718^{-\lambda \; 3 \text{s}} = 600
\tag{3}
\label{eq:probrelat-fr}
\end{equation}

Nous pouvons manipuler le côté gauche de cette équation pour montrer

\begin{equation}
N(t = 3 \text{ s}) = 2.718^{-\lambda \; 3 \text{s}} = \frac{600}{1000} = 0.6 = 60 \%
\end{equation}

Où si nous substituons notre nouvelle expression pour $N(t = 3 \text{ s})$ à l'équation \ref{eq:probrelat-fr}, nous voyons que cette fonction est analogue à une probabilité. En d'autres termes, l'expression

\begin{equation}
2.718^{-\lambda t}
\end{equation}

est la probabilité qu'un atome <**> ne soit pas <**> décroissant à un instant donné $t$, ou

\begin{equation}
P_{exists}(t) = 2.718^{-\lambda t}
\label{eq:exist-fr}
\tag{4}
\end{equation}

Où $P_{exists}(t)$ est la probabilité qu'un atome existe à l'instant $t$.


Maintenant que nous avons une expression pour la probabilité qu'un atome <**> ne soit pas <**> décroît, nous pouvons maintenant calculer la probabilité qu'un atome <**> soit <**> décroissant. La probabilité qu'un atome <**> soit <**> décroît est simplement une moins la probabilité qu'il ne se désintègre pas, ou

\begin{equation}
P_{decayed}(t) = 1 - P_{exists}(t) = 1 - 2.718^{-\lambda t}
\label{eq:decayed-fr}
\tag{5}
\end{equation}

Où $P_{decayed}(t)$ est la probabilité qu'un atome ait décru par le temps $t$.

Cela signifie que l'expression dans l'équation \ref{eq:decay-fr} est une "idéalisation lisse". La désintégration radioactive est un processus statistique, et les données réelles ne_ approchent que <_> la solution de l’équation \ref{eq:decay-fr}. En réalité, nos observations de radioactif fluctueront au hasard autour de cette solution. Pour visualiser ceci, nous pouvons utiliser Python pour avoir une idée de ce que nous pouvons mesurer si nous observons une population réelle de désintégration d'atomes.

## Exemple Python

### La désintégration radioactive en analogie avec le lancement d'une pièce biaisée

Supposons que vous passiez en revue un sac d'atomes et que vous décidiez quels sont ceux qui se décomposent en les jetant selon leur demi-vie. Supposons qu'à votre époque actuelle, chaque atome ait une probabilité de décroissance de 50%. En faisant cela manuellement, vous pouvez peut-être choisir un atome et lancer une pièce pour décider de son sort. Si la pièce montre des têtes, vous pouvez garder l’atome (l’atome ne se désintègre pas), et si la pièce présente des queues, vous pouvez la jeter (l’atome se désintègre). Une fois que vous avez parcouru l'ensemble des atomes, vous aurez jeté ces atomes au hasard en fonction de leur probabilité de pourriture. C'est une méthodologie raisonnable: vous supprimez environ la moitié des atomes au hasard. De plus, le fait de retirer un atome est totalement indépendant de l’action de retirer un atome. Y a-t-il un moyen de généraliser ce processus et de l'appliquer à ce que nous avons appris sur la désintégration nucléaire?
 
Heureusement, Python permet de dessiner des nombres aléatoires uniformément répartis entre 0 et 1 algorithmiquement. En d'autres termes, il existe une fonction Python qui nous permet de tirer au hasard un pourcentage compris entre 0 et 100%, chaque pourcentage pouvant également être dessiné. En utilisant cette fonction, nous pouvons comparer notre nouveau pourcentage aléatoire à nos probabilités de décroissance pour décider si notre atome se désintègre ou non! C'est exactement la même chose que de retourner une pièce pour chaque atome. Sauf que maintenant, nous utilisons une "pièce biaisée" qui peut favoriser la tête par-dessus la queue (ou dans notre cas, se décomposer ou ne pas se détériorer).

### Simuler la désintégration nucléaire
La manière dont nous examinons notre "sac" d'atomes peut être décrite comme un organigramme, comme illustré ci-dessous.
>![alt text](images/nuclear_decay.png)
> Nous voyons ici l’organigramme du processus de simulation de la désintégration nucléaire. Nous voyons ici que pour chaque instance dans le temps $t_i$ nous avons un certain nombre d'atomes $N$. À ce moment précis, nous comparons la probabilité de décroissance de chaque atome à un nombre aléatoire unique $r$. Si le nombre aléatoire est inférieur à notre probabilité de décroissance, alors on dit que l'atome s'est désintégré. Nous retirons ensuite cet atome du sac. Une fois que nous vérifions chaque atome à une instance particulière dans le temps, nous passons à notre instance suivante avec un nouveau nombre d'atomes $N$ et répétons le processus

Le pseudo-code ci-dessous montre comment écrire ce processus avec du code. Ici, nous regardons à travers notre "sac" d'atomes, puis la comparaison entre le nombre aléatoire et notre probabilité de décroissance calculée agit comme un tirage au sort (biaisé) pour décider si l'atome se désintègre ou non.


```python

for each instance_in_time in time:
    probability_of_decay = 1 - exp(-lambda * instance_in_time)
    for each atom in atoms:
        # This is a random number between 0 and 1
        random_percent = random_number()
        if prob_decay > r:
            atom.decays()
```
    
Où ici `instance _dans_ time` est un pas de temps également espacé (pas de seconde par exemple), et nous parcourons plusieurs pas de temps. À chaque pas de temps, nous calculons alors la probabilité qu'un atome se désintègre à cette instance dans le temps par l'équation 5. Nous parcourons ensuite tous les atomes, générons un nombre aléatoire compris entre 0 et 1 (pourcentage aléatoire compris entre 0 et 100). Ce nombre est ensuite utilisé pour décider si un atome se désintègre en atome. Ensuite, nous le faisons à nouveau au prochain pas. Pour ce qui est de l'analogie avec les pièces, si `probabilité _de_ decay` est supérieur à` random _percent`, notre pièce est à la tête et l'atome ne se désintègre pas. Si `probabilité_ de _decay` est inférieur à` random_ percent`, notre tirage au sort se fait en queues et l’atome se désintègre.

Pour illustrer cela avec un exemple, si `probabilité _de_ decay` est de 76% et que le nombre aléatoire` random _percent` était de 45%, notre atome ne se désintégrerait pas. Si `random_ percent` était supérieur à 76%, l'atome décroît. Comme les nombres aléatoires sont uniformément distribués <_>, cela signifie que chaque pourcentage entre 0 et 100 est tout aussi probable.
    
Notez que nous calculons la probabilité de décroissance **avant** nous entrons dans la boucle où nous voyons si un atome se désintègre ou non. En effet, à un instant donné, la probabilité de décroissance pour chaque atome est **indépendante**. Si un atome se désintègre, cela n'affecte pas si l'un des autres décroît également.

Ce qu'il faut garder à l'esprit, c'est que la base de cette simulation est d'une simplicité trompeuse. Nous décidons simplement **si** chaque atome décroît en retournant une pièce biaisée à chaque étape. Si vous deviez le faire "à la main", vous seriez littéralement en train de lancer des pièces et de compter des atomes. Et c'est précisément ce que fait notre simulation, plus rapidement que possible.

````python
from IPython.display import HTML
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt 
import random
import pandas as pd
from ipywidgets import interactive
import time
from IPython import display
import plotly as py 
import plotly.graph_objs as go
py.offline.init_notebook_mode()

%matplotlib inline
hide_me = ''
HTML('''<script>
code_show=true; 
function code_toggle() {
  if (code_show) {
    $('div.input').each(function(id) {
      el = $(this).find('.cm-variable:first');
      if (id == 0 || el.text() == 'hide_me') {
        $(this).hide();
      }
    });
    $('div.output_prompt').css('opacity', 0);
  } else {
    $('div.input').each(function(id) {
      $(this).show();
    });
    $('div.output_prompt').css('opacity', 1);
  }
  code_show = !code_show
} 
$( document ).ready(code_toggle);
</script>
<form action="javascript:code_toggle()"><input style="opacity:0" type="submit" value="Click here to toggle on/off the raw code."></form>''')
````

````python
hide_me 


def HalfLifeEquation(N_0, lamb,t):
    return N_0 * 2.71828 ** (-lamb * t)

# Halflife of carbon-15 is 2.249 seconds

t = np.linspace(0,4*2.449,500)
plt.figure(figsize=(3, 2), dpi= 175, facecolor='w', edgecolor='k')
plt.plot(t, HalfLifeEquation(1000, np.log(2)/2.249, t), label="Analytic")
plt.ylabel("Number of atoms")
plt.xlabel("t (seconds)")
plt.title("Carbon - 15")


def Decay(N_0, lamb,max_time, steps=100):
    time_atom_pairs = []
    N = N_0
    time_step = max_time/steps
    t = 0

    for i in range(steps):
        
        # For each step in time, we need to check if each atom 
        # decays or not note we only calculate our probability
        # once because at a single instant in time, the probability
        # of decay is the same for each -- they are independent 
        # events. 
        
        # this is the equation we saw eariler 
        # Divide by N to reuse function
        p = 1 - HalfLifeEquation(N, lamb, time_step)/N
        # Maybe explain the _ part? 
        for _ in range(N):
            
            # This is a 'random' number between 0 and 1
            # that we will use to compare to our decay probability
            r = random.uniform(0,1)
           
            # if our decay probability for the atom is greater, that 
            # atom decays. 
            if r < p:
                N = N -1
           
            # if we run out of atoms we should stop
            if N == 0:
                break
        
        # We have now moved forward in time once again!
        t = t + time_step
        # Store how many atoms at a given time t we have in 
        # an array. 
        time_atom_pairs.append([N, t])
        
    return time_atom_pairs

decays = Decay(1000, np.log(2)/2.249, 10, steps=500)



y,x = np.array(decays).T
#print(x)
#print(decays)
plt.plot(x,y, label = "Simulated")
plt.legend()
plt.show()
            
        
            
    
HTML('''<script>
    code_show=true; 
    function code_toggle() {
        if (code_show){
            $('div.cell.code_cell.rendered.selected div.input').hide();
        } else {
            $('div.cell.code_cell.rendered.selected div.input').show();
        }
        code_show = !code_show
    } 

    $( document ).ready(code_toggle);
    </script>

    To show/hide the algorithm which generates the plot above, click <a href="javascript:code_toggle()">here</a>.''')
   
````

# Processus de fusion et origine des éléments lourds

Il est bien accepté que chaque étoile dans l'univers observable est principalement alimentée par la fusion de l'hydrogène en hélium. Les étoiles gagnent également de l'énergie à partir de la fusion d'éléments plus lourds jusqu'à la mathématique du fer 56. En effet, la quantité d'énergie qu'il faut pour combiner des éléments passé ce point fait [nécessite plus d'énergie qu'ils produisent](https://fr.wikipedia.org/wiki/Liaison_nucléaire). Généralement, lorsqu'une étoile commence à créer du fer, l'étoile est en fin de vie. Si les étoiles cessent de produire les éléments au fer 56, comment obtient-on le reste du tableau périodique? Eh bien, la nucléosynthèse stellaire!

Nucléosynthèse stellaire

[Nucléosynthèse stellaire](https://fr.wikipedia.org/wiki/Nucléosynthése_stellaire) est la création de nucléides atomiques lors d'événements astrophysiques à haute énergie tels que [Supernova](https://fr.wikipedia.org/wiki/Nucléosynthése_explosive), [les fusions étoiles à neutrons](https://physics.aps.org/articles/v10/114) ou [quark novae](http://www.quarknova.ca/_include/files/feature_quark_stars.pdf). Au cours de ces événements à haute énergie, il est théorisé qu’il existe un nombre incalculable de neutrons libres pouvant être liés à divers noyaux. Un atome se déplaçant dans un grand nombre de neutrons ramasserait de nombreux neutrons en créant rapidement des isotopes lourds et instables.
Ce processus de noyaux absorbant un neutron libre est appelé capture de neutrons. La capture de neutrons a la formule générale pour un atome $X$ et un neutron $N$

\begin{equation}
^{A}_ZX^N + N \rightarrow ^{A+1}_Z X^{N+1}
\end{equation}

Où cela a formé un isotope ** différent ** de l'atome $X$.
Dans un cas astrophysique de haute énergie, le procédé de collecte des noyaux rapidement neutrons est connu sous le nom ** ** $r$-processus ou [**processus de capture de neutrons rapides**](https://fr.wikipedia.org/wiki/Processus_r). Le processus $r$ est responsable de la création de nombreux éléments plus lourds que le fer. Il existe également un autre processus connu sous le nom de [**processus de capture de neutrons lents**](https://fr.wikipedia.org/wiki/Processus_s), ou le ** $s$-process **. La différence entre les processus $r$ et $s$, pour notre propos, est mineure et la différence entre les deux n'est que dans la rapidité avec laquelle ils se produisent.

Une question naturelle à poser est "Comment cela crée-t-il la grande variété d'éléments? Cela ne fait que me dire qu'il devrait y avoir des isotopes incroyablement riches en neutrons!". Cela est très vrai, cependant, l'autre moitié des processus $r$ et $s$ est qu'ils créent des isotopes incroyablement riches_instables_riches en neutrons. Et que font les isotopes instables? Eh bien, ils $\beta$ décadence bien sûr! Dans ce cas, l'excès de neutrons se désintègre en émettant un électron et en se transformant en proton, créant ainsi un nouvel élément! Bien sûr, si cela se produit trop, il y a aussi une possibilité (beaucoup plus rare) qu'un proton émette un positron devenant ainsi un neutron. Ces processus sont décrits pour des atomes arbitraires $X$, $Y$ et $W$ ci-dessous, premier $\beta^-$ decay

\begin{equation}
^{A}_ZX^N  \rightarrow ^{A}_{Z+1} Y^{N-1} + e^- + \bar{\nu}
\end{equation}

et puis $\beta^+$ se désintégrer

\begin{equation}
^{A}_ZX^N  \rightarrow ^{A}_{Z-1} W^{N+1} + e^+ + \nu
\end{equation}


L'idée de la nucléosynthèse stellaire est que les noyaux rassemblent les neutrons en excès et ensuite la décomposition mathidique en stabilité. Au cours d'un événement de nucléosynthèse stellaire, de nombreux autres processus se produisent simultanément, ce qui complique le processus. Ces événements principaux dans les processus $r$ et $s$ sont suffisants pour la démonstration suivante.

## Un modèle (très) simplifié

Pour obtenir une compréhension de base du cycle de vie d'un nucléon pris au piège au milieu de la $r$ et $s processes, we're going to construct a toy model<sup>[1](#footnote1)</sup> of stellar nucleosynthesis that should produce some interesting visuals.




First, start with a with a generic atom, any old atom will do, and we'll specify and arbitrary number of neutrons $N$ and an arbitrary number of protons $Z$. Once the atom is specified,  assume it is trapped inside the $r$ or $s$ process. Then we will toss a coin to decide if our nuclei captures a neutron, or beta decays. If it captures a neutron we'll increase the neutron count by one, if not we'll beta decay and lose a neutron and gain a proton, outlined in the flowchart below

>![alt text](images/proton_decay_.png)

> The image above depicts the calculation process required to create our toy model of nucleosynthesis. The green box outlines the calculation process when protons are not allowed to decay. The gray box outlines the process when protons are allowed to decay. To begin, we have an initial number of neutrons $N$ and an initial number if protons $Z$. We then draw a random number $r$ uniformly from the range $ [0,1] $, and compare that to our neutron capture probability $p_{} NC $. If $r$ is greater than the probability of neutron capture, we gain a neutron, if not, we beta decay thereby losing a neutron and gaining a proton. Alternatively, if we allow proton decay, if our random number has dictated that we beta decay, we then draw a second random number $r_2$ uniformly from $ [0,1] $ and compare it to our probability of proton decay $p_{PD} $. This comparison then decides which mode of decay that our atom will undergo. We note that the exit condition in the widget you'll soon see is slightly different preventing cases where reaching $Z / N> 1$ may take a long time, or is changed to $N / Z> 1 $ s'il y a plus de protons que de neutrons au début de la simulation.

````python
hide_me
# This is the exit condition of our loop, scaled to the original
# number of neutrons and protons
def Condition(Z,N, Z_0, N_0):
    if Z_0 >= N_0:
        return N/Z
    if N_0 > Z_0:
        return Z/N
        
# This is our toy model 

def SimulateNewElement(N,Z, BetaNegative=False, ProbNegative = 0.1):
    # Initialize some lists and counters
    count = 0
    NArray = []
    ZArray = []
    min_iterations = 20
    N_0 = N
    Z_0 = Z
    
    # Note that this is an infinite loop, we're relying on 
    # the break function to escape based on the conditions we've 
    # decided on. 
   
    while True:
        # Count how many times we've cycled and draw a random number
        count += 1
        r = random.uniform(0,1)
        # neutron capture
        if r >= 0.5:
            N = N + 1
            NArray.append(N)
            ZArray.append(Z)
        # beta decay
        elif r < 0.5:
            # draw a second random number to decide what kind of beta decay
            r2 = random.uniform(0,1)
            # Can only beta decay if we have a neutron and a proton
            if N > 0 and Z > 0:
                # IF we're allowing proton decay
                if BetaNegative:
                    # Beta minus decay
                    if r2 < 1 - ProbNegative:
                        N = N - 1 
                        Z = Z + 1
                    # beta plus decay
                    else:
                        N = N + 1
                        Z = Z - 1
                else:
                    # beta minus decay
                    N = N - 1
                    Z = Z + 1
            # save our counts at each iteration
            NArray.append(N)
            ZArray.append(Z)
        # Decide if we should leave
        if Condition(Z,N,Z_0,N_0) > 1.01 and count > min_iterations:
            if count > min_iterations:
                print("Reached Z = N line and passed minimum iterations, simulation finished")
            else:
                print("Ran minimum number of times and passed Z=N line.")
            break 
        if count > 200:
            print("Taking too long, exiting early after 200 iterations")
            print("Try lowering the probability of negative beta decay")
            print("Or just simply running this simulation again")
            break
    
            
    return np.array(NArray), np.array(ZArray), np.linspace(0,count,count)


HTML('''<script>
    code_show=true; 
    function code_toggle() {
        if (code_show){
            $('div.cell.code_cell.rendered.selected div.input').hide();
        } else {
            $('div.cell.code_cell.rendered.selected div.input').show();
        }
        code_show = !code_show
    } 

    $( document ).ready(code_toggle);
    </script>

    To show/hide the algorithm that generates our toy model
    of nucleosynthesis, click <a href="javascript:code_toggle()">here</a>.''')

    

            
````
